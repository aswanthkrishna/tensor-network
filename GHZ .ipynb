```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

```python
no_of_qubits = 5
qc = qiskit.QuantumCircuit(no_of_qubits,no_of_qubits)
qc.h(0)
for i in range(1,no_of_qubits):
    qc.cx(0,i)
for i in range(0,no_of_qubits):
    qc.measure(i,i)
%matplotlib inline
qc.draw(output = 'mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc,simulator)
result = job.result()
plot_histogram(result.get_counts())
```
